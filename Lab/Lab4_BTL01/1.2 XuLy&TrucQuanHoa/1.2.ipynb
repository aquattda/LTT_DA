# 1.2.1.  Ôn tập Lý thuyết xử lý và trực quan hóa dữ liệu 
## Câu hỏi 1: Trực quan hóa dữ liệu có vai trò gì trong phân tích dữ liệu và tại sao nó quan trọng trong giai đoạn khám phá dữ liệu (EDA)?

    Vai trò: Trực quan hóa dữ liệu (Data Visualization) biến những con số khô khan thành hình ảnh, giúp con người nhận diện xu hướng, phát hiện các điểm bất thường, và hiểu các mối quan hệ trong dữ liệu một cách nhanh chóng.

    Tầm quan trọng trong EDA (Exploratory Data Analysis): Đây là công cụ thiết yếu trong giai đoạn khám phá dữ liệu vì nó cho phép chúng ta "nhìn" vào cấu trúc của dữ liệu, đặt ra các câu hỏi và định hướng cho các bước phân tích sâu hơn mà không cần đến các mô hình phức tạp.

## Câu hỏi 2: Các loại biểu đồ phổ biến (như histogram, scatter plot, boxplot, bar chart) được sử dụng trong các trường hợp nào?

    Histogram: Dùng khi muốn xem phân bố của một biến số duy nhất (ví dụ: phân bố tuổi của khách hàng).

    Scatter Plot (Biểu đồ phân tán): Dùng khi muốn kiểm tra mối quan hệ giữa hai biến số (ví dụ: mối liên hệ giữa chi tiêu quảng cáo và doanh thu).

    Boxplot (Biểu đồ hộp): Dùng khi muốn so sánh phân bố của một biến số qua nhiều nhóm khác nhau và phát hiện outliers (ví dụ: so sánh mức lương giữa các phòng ban).

    Bar Chart (Biểu đồ cột): Dùng khi muốn so sánh số lượng hoặc một giá trị tổng hợp (như trung bình) của các nhóm (dữ liệu phân loại) khác nhau (ví dụ: so sánh doanh số bán hàng theo từng khu vực).

## Câu hỏi 3: Làm thế nào để chọn loại biểu đồ phù hợp với đặc điểm của dữ liệu?

Việc lựa chọn biểu đồ phụ thuộc vào câu hỏi bạn muốn trả lời và loại dữ liệu bạn có:

    Dữ liệu phân loại (Categorical): Dùng Bar Chart để so sánh các nhóm, hoặc Pie Chart (biểu đồ tròn) để thể hiện tỷ lệ phần trăm.

    Dữ liệu số (Numerical):

        Để xem phân bố của một biến: Dùng Histogram hoặc Boxplot.

        Để xem mối quan hệ giữa hai biến: Dùng Scatter Plot.

    Dữ liệu thời gian (Time-Series): Dùng Line Chart (biểu đồ đường) để theo dõi sự thay đổi của một giá trị theo thời gian.

## Câu hỏi 4: Sự khác biệt giữa các thư viện trực quan hóa trong Python như Matplotlib, Seaborn và Plotly là gì?

    Matplotlib: Là thư viện nền tảng, có khả năng tùy biến cực kỳ cao. Giống như bạn tự xây một ngôi nhà từ móng, bạn kiểm soát mọi chi tiết nhưng cần nhiều công sức hơn.

    Seaborn: Được xây dựng dựa trên Matplotlib. Cung cấp các biểu đồ thống kê đẹp mắt và phức tạp hơn với ít mã hơn. Giống như sử dụng một bộ khung nhà dựng sẵn, nhanh và đẹp nhưng ít linh hoạt hơn Matplotlib.

    Plotly: Nổi bật với khả năng tạo các biểu đồ tương tác (interactive), cho phép bạn di chuột để xem thông tin, phóng to, thu nhỏ. Rất phù hợp để xây dựng dashboard hoặc trình bày trên web.

## Câu hỏi 5: Những nguyên tắc thiết kế nào cần tuân thủ để tạo ra một biểu đồ trực quan hóa dễ hiểu và hiệu quả?

    Rõ ràng và Đơn giản: Loại bỏ các yếu tố không cần thiết (đường lưới dày, màu sắc lòe loẹt). Mục tiêu là làm nổi bật dữ liệu.

    Ghi nhãn đầy đủ: Luôn có tiêu đề cho biểu đồ, nhãn cho các trục (kèm theo đơn vị nếu có).

    Chọn đúng loại biểu đồ: Sử dụng loại biểu đồ phù hợp nhất với thông điệp bạn muốn truyền tải.

    Sử dụng màu sắc có chủ đích: Dùng màu để làm nổi bật thông tin quan trọng hoặc để phân biệt các nhóm, tránh dùng quá nhiều màu gây rối mắt.

    Đảm bảo tính trung thực: Trục Y phải bắt đầu từ 0 (đặc biệt với biểu đồ cột) để tránh gây hiểu lầm về tỷ lệ.

## Câu hỏi 6: Các bước lý thuyết để tạo một biểu đồ đơn giản là gì?

Để tạo một biểu đồ như histogram bằng thư viện Matplotlib, về mặt lý thuyết, bạn cần thực hiện các bước sau:

    Chuẩn bị dữ liệu bạn muốn vẽ.

    Gọi hàm vẽ biểu đồ tương ứng (ví dụ: plt.hist() cho histogram) và truyền dữ liệu vào.

    Tùy chỉnh biểu đồ bằng cách thêm tiêu đề, nhãn cho các trục để làm rõ ý nghĩa.

    Cuối cùng, hiển thị biểu đồ ra màn hình hoặc lưu thành một tệp tin.

## Câu hỏi 7: Làm thế nào để xuất biểu đồ ra các định dạng file khác nhau?

Hầu hết các thư viện trực quan hóa trong Python đều cung cấp một hàm chuyên dụng để lưu biểu đồ. Ví dụ, trong Matplotlib, bạn có thể sử dụng hàm savefig(). Bạn chỉ cần cung cấp tên tệp và định dạng mong muốn (như PNG, PDF, SVG) cho hàm này. Lệnh lưu tệp thường được thực hiện ngay trước lệnh hiển thị biểu đồ.

### Import libraries

```python
import numpy as np
import pandas as pd
from scipy import stats # comment
import matplotlib.pyplot as plt
```

### Load dataset

```python
houseprices_data = pd.read_csv("HousingPrices-Amsterdam-August-2021.csv")
```

### 1. Chuẩn bị dữ liệu cho trực quan hóa dữ liệu

```python
houseprices_data = houseprices_data[['Zip', 'Price', 'Area', 'Room']]
# Create a PriceperSqm variable based on the Price and Area variables:
houseprices_data['PriceperSqm'] = houseprices_data['Price']
houseprices_data['Area']
```

### 2. Trực quan hóa dữ liệu với thư viện Matplotlib

```python
houseprices_sorted = houseprices_data.sort_values('Price', ascending =
False)
houseprices_sorted.head()
# case 1: basic
plt.figure(figsize= (12,6))
x = houseprices_sorted['Zip'][0:10]
y = houseprices_sorted['Price'][0:10]
plt.bar(x,y)
plt.show()
# case 2: advanced 1
plt.figure(figsize= (12,6))
plt.bar(x,y)
plt.title('Top 10 Areas with the highest house prices', fontsize=15)
plt.xlabel('Zip code', fontsize = 12)
plt.xticks(fontsize=10)
plt.ylabel('House prices in millions', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
# case 3: advanced 2
fig, ax = plt.subplots(figsize=(40,18))
x = houseprices_sorted['Zip'][0:10]
y = houseprices_sorted['Price'][0:10]
y1 = houseprices_sorted['PriceperSqm'][0:10]
plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(fontsize=17)
plt.ylabel('House prices in millions', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices',
fontsize=25)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=17)
plt.ylabel('House prices per sqm', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Areas with the highest house prices per sqm',
fontsize=25)
plt.show()
```

### 3. Trực quan hóa dữ liệu với thư viện Seaborn

```python
import seaborn as sns
# case 1: basic
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
sns.barplot(data= data, x= 'Zip',y = 'Price')
# case 2: advanced 1
plt.figure(figsize= (12,6))
data = houseprices_sorted[0:10]
ax = sns.barplot(data= data, x= 'Zip',y = 'Price')
ax.set_xlabel('Zip code',fontsize = 15)
ax.set_ylabel('House prices in millions', fontsize = 15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize= 20)
# case 3: view multiple perspectives at once
fig, ax = plt.subplots(1, 2,figsize=(40,18))
data = houseprices_sorted[0:10]
sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Zip',y = 'Price', ax = ax[0])
ax1.set_xlabel('Zip code')
ax1.set_ylabel('House prices in millions')
ax1.set_title('Top 10 Areas with the highest house prices')
ax2 = sns.barplot(data= data, x= 'Zip',y = 'PriceperSqm',
ax=ax[1])
ax2.set_xlabel('Zip code')
ax2.set_ylabel('House prices per sqm')
ax2.set_title('Top 10 Areas with the highest price per sqm')
```

### 2.3. Thay đổi tên cột (Columns name change)

```python
# winequality_df = winequality_df.rename(columns={"fixed acidity": "fixed_acidity"}) #single

cols_to_fix = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "free sulfur dioxide", "total sulfur dioxide"]
def normalize_selected_cols(df, cols):
    mapping = {c:c.strip().lower().replace(" ", "_") for c in cols}
    return df.rename(columns = mapping)

winequality_df = normalize_selected_cols(winequality_df, cols_to_fix)
winequality_df.head()
```
