# 1.3.1.  Ôn tập Phân tích đơn biến và hai biến
## Câu hỏi 1: Phân tích đơn biến (univariate analysis) là gì? Nó khác gì với phân tích hai biến (bivariate analysis)?

    Phân tích đơn biến: Là việc phân tích từng biến một một cách riêng lẻ. Mục tiêu là để mô tả, tóm tắt và tìm ra các đặc điểm của duy nhất biến đó (ví dụ: tính giá trị trung bình của cột "tuổi").

    Phân tích hai biến: Là việc phân tích mối quan hệ giữa hai biến cùng một lúc. Mục tiêu là để xem chúng có tương quan, ảnh hưởng lẫn nhau như thế nào (ví dụ: xem xét mối quan hệ giữa "chiều cao" và "cân nặng").

## Câu hỏi 2: Các thước đo thống kê nào thường được sử dụng trong phân tích đơn biến?

Các thước đo phổ biến nhất được sử dụng trong phân tích đơn biến bao gồm:

    Các thước đo xu hướng trung tâm: Trung bình (mean), trung vị (median), và yếu vị (mode) để xác định giá trị "điển hình" của biến.

    Các thước đo độ phân tán: Độ lệch chuẩn (standard deviation), phương sai (variance), phạm vi (range), và khoảng tứ phân vị (IQR) để đo lường mức độ trải rộng của dữ liệu.

## Câu hỏi 3: Trong phân tích hai biến, làm thế nào để xác định mối quan hệ giữa hai biến?

Để xác định mối quan hệ giữa hai biến, chúng ta thường sử dụng:

    Trực quan hóa: Dùng các biểu đồ như Scatter Plot để "nhìn" xem có xu hướng nào giữa hai biến số hay không (ví dụ: chúng cùng tăng hay cùng giảm).

    Đo lường thống kê: Dùng các hệ số như Tương quan (Correlation) để đo lường sức mạnh và chiều hướng của mối quan hệ tuyến tính.

    Lưu ý: Việc xác định mối quan hệ nhân quả (causation) phức tạp hơn nhiều và đòi hỏi các phương pháp thử nghiệm hoặc phân tích sâu hơn, không thể chỉ kết luận từ việc quan sát dữ liệu.

## Câu hỏi 4: Sự khác biệt giữa tương quan (correlation) và hiệp phương sai (covariance) là gì?

    Hiệp phương sai (Covariance): Chỉ cho biết chiều hướng của mối quan hệ giữa hai biến.

        Giá trị dương: Hai biến có xu hướng cùng tăng.

        Giá trị âm: Một biến tăng thì biến kia có xu hướng giảm.

        Nó không được chuẩn hóa, nên rất khó để so sánh mức độ mạnh yếu giữa các cặp biến khác nhau.

    Tương quan (Correlation): Vừa cho biết chiều hướng, vừa cho biết sức mạnh của mối quan hệ tuyến tính.

        Giá trị của nó luôn nằm trong khoảng từ -1 đến 1, đã được chuẩn hóa.

        Giá trị gần 1 hoặc -1 cho thấy mối quan hệ mạnh, gần 0 cho thấy mối quan hệ yếu.

## Câu hỏi 5: Khi nào nên sử dụng biểu đồ trong phân tích đơn biến so với phân tích hai biến?

    Phân tích đơn biến: Sử dụng biểu đồ để hiểu phân bố của một biến.

        Ví dụ: Dùng Histogram hoặc Boxplot để xem dữ liệu tập trung ở đâu, có đối xứng hay bị lệch không, và có outliers hay không.

    Phân tích hai biến: Sử dụng biểu đồ để khám phá mối quan hệ giữa hai biến.

        Ví dụ: Dùng Scatter Plot để tìm mối quan hệ tuyến tính giữa hai biến số. Dùng Boxplot để so sánh sự phân bố của một biến số qua các nhóm của một biến phân loại.

## Câu hỏi 6: Đoạn code mẫu để tạo biểu đồ scatter plot hoặc heatmap để phân tích mối quan hệ giữa hai biến?

Dưới đây là các đoạn code mẫu sử dụng thư viện matplotlib và seaborn trong Python.

    Scatter Plot (Biểu đồ phân tán) với Matplotlib:

        Biểu đồ này rất hữu ích để xem mối quan hệ giữa hai biến số.

    import matplotlib.pyplot as plt
    import numpy as np

    # Tạo dữ liệu mẫu
    x_data = np.random.rand(50) * 10
    y_data = 2 * x_data + np.random.randn(50) * 2

    # Vẽ biểu đồ scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(x_data, y_data, alpha=0.7)

    # Thêm tiêu đề và nhãn
    plt.title('Biểu đồ phân tán ví dụ')
    plt.xlabel('Biến X')
    plt.ylabel('Biến Y')
    plt.grid(True)
    plt.show()

    Heatmap (Bản đồ nhiệt) với Seaborn:

        Heatmap thường được dùng để trực quan hóa ma trận tương quan giữa nhiều biến, giúp nhanh chóng phát hiện các biến có tương quan mạnh.

    import seaborn as sns
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    # Tạo dữ liệu mẫu và DataFrame
    data = np.random.rand(10, 5)
    df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

    # Tính toán ma trận tương quan
    correlation_matrix = df.corr()

    # Vẽ heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

    # Thêm tiêu đề
    plt.title('Bản đồ nhiệt ma trận tương quan')
    plt.show()

## Câu hỏi 7: Làm thế nào để trực quan hóa mối quan hệ giữa một biến số và một biến phân loại?

Để trực quan hóa mối quan hệ này, bạn có thể dùng các biểu đồ so sánh phân bố:

    Boxplot (Biểu đồ hộp): Vẽ một biểu đồ hộp riêng cho biến số ứng với mỗi nhóm của biến phân loại. Điều này cho phép bạn so sánh trực quan các giá trị trung vị, độ phân tán (IQR), và outliers giữa các nhóm.

    Violin Plot (Biểu đồ Violin): Tương tự như Boxplot nhưng nó còn hiển thị cả mật độ phân bố của dữ liệu ở mỗi bên. Nó cung cấp một cái nhìn chi tiết hơn về hình dạng phân bố của dữ liệu trong từng nhóm.

## Nhiệm vụ 1: phân tích dữ liệu đơn biến trên dữ liệu về chim cánh cụt

### 1. Import thư viện và nạp dữ liệu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins_data = pd.read_csv("penguins_lter.csv")
penguins_data = penguins_data[['Species','Culmen Length (mm)']]
```

### 2. Phân tích đơn biến bằng Histogram

```python
# case 1: basic
sns.histplot( data = penguins_data, x= penguins_data["Culmen Length (mm)"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.histplot(data = penguins_data, x = penguins_data["Culmen Length (mm)"])
ax.set_xlabel('Culmen Length in mm',fontsize = 15)
ax.set_ylabel('Count of records', fontsize = 15)
ax.set_title('Univariate analysis of Culmen Length',fontsize= 20)
```

### 3. Phân tích đơn biến bằng bar chart

```python
# case 1: basic
sns.countplot(data = penguins_data, x= penguins_data['Species'])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.countplot(data = penguins_data, x= penguins_data['Species'])
ax.set_xlabel('Penguin Species',fontsize = 15)
ax.set_ylabel('Count of records',fontsize = 15)
ax.set_title('Univariate analysis of Penguin Species', fontsize= 20)
ax.set_title('Univariate analysis of Culmen Length', fontsize= 20)
```

### 4. Phân tích đơn biến bằng biểu đồ tròn (Pie-chart)

```python
penguins_group = penguins_data.groupby('Species').count()
penguins_group = penguins_group.reset_index()
# case 1: basic
plt.pie(penguins_group["Culmen Length (mm)"], labels =
penguins_group['Species'])
plt.show()
# case 2: advanced
cols = ['g', 'b', 'r']
plt.pie(penguins_group["Culmen Length (mm)"], labels =
penguins_group['Species'],colors = cols)
plt.title('Univariate Analysis of Species', fontsize=15)
plt.show()
```

## Nhiệm vụ 2: Phân tích dữ liệu đơn biến trên dữ liệu giá nhà

### 1. Import thư viện, nạp dữ liệu giá nhà và phân tích đơn biến dựa vào boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
houseprices_data = pd.read_csv("HousingPrices-Amsterdam-August-2021.csv")
houseprices_data = houseprices_data[['Zip','Price','Area','Room']]
# case 1: Create a boxplot using the boxplot method
sns.boxplot(data = houseprices_data, x= houseprices_data["Price"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.boxplot(data = houseprices_data, x= houseprices_data["Price"])
ax.set_xlabel('House Prices in millions',fontsize = 15)
ax.set_title('Univariate analysis of House Prices', fontsize= 20)
plt.ticklabel_format(style='plain', axis='x')
```

### 2. Phân tích dữ liệu đơn biến dựa vào violin plot

```python
# case 1: basic
sns.violinplot(data = houseprices_data, x= houseprices_data["Price"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.violinplot(data = houseprices_data, x=houseprices_data["Price"])
ax.set_xlabel('House Prices in millions',fontsize = 15)
ax.set_title('Univariate analysis of House Prices', fontsize= 20)
plt.ticklabel_format(style='plain', axis='x')
```
